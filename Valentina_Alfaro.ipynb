Inserto librerías:

```python
import matplotlib.pyplot as plt
import numpy as np
```

Declaro las constantes

```python
h = 6.626e-34
c = 3.0e+8
k = 1.38e-23
T = 0.008
p = np.pi
```

Declaro las constantes operadas para la simplificación de la función

```python
a = 2*h/(c**2)
b = h/k*T
c = 2*k*T/(c**2)
```

Declaro la variable de la frecuencia de onda en metros

```python
frecuencia = np.arange(1e7, 1e9, 1e3) 
```

Definición de la primera función B(v,T)

```python
def funcion1(frecuencia):
    valores = a*pow(frecuencia,3)/(np.exp(b*frecuencia)-1)
    return valores 
```

Definición de la segunda función B_v(T)

```python
def funcion2(frecuencia):
    valores = a*pow(frecuencia,3)*np.exp(-b*frecuencia)
    return valores 
```

Definición de la tercera función B(T)

```python
def funcion3(frecuencia):
    valores = c*(frecuencia**2)
    return valores 
```

Se crean los datos para las tres funciones de aproximación

```python
curva1 = funcion1(frecuencia)
curva2 = funcion2(frecuencia)
curva3 = funcion3(frecuencia)
```

Se proceden a graficar las curvas obtenidas con sus etiquetas

```python
plt.plot(frecuencia, curva1, '-r',label='Plank') 
plt.plot(frecuencia, curva2, '-g',label='Wien') 
plt.plot(frecuencia, curva3, '-b',label='Rayleigh-Jeans') 
plt.legend(loc='best')
plt.loglog(frecuencia,curva1)
plt.loglog(frecuencia,curva2)
plt.loglog(frecuencia,curva3)
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Intensidad Luminosa [J m^2]')
```
